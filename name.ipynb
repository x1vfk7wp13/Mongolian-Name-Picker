```python
names = []

with open("names.txt", "r", encoding="utf-8") as file:
    for line in file:
        # Split by whitespace and filter out single-letter names
        words = line.strip().split()
        filtered = [name for name in words if len(name) > 1]
        names.extend(filtered)

print(names)
```

4 letter names

```python
names = []

with open("names.txt", "r", encoding="utf-8") as file:
    for line in file:
        words = line.strip().split()
        # Keep only names that are exactly 4 characters long
        four_letter_names = [name for name in words if len(name) == 4]
        names.extend(four_letter_names)

print(names)
```

```python
import csv

# Path to your CSV file
file_path = "mongolian_personal_names.csv"

# Open and parse the CSV file
with open(file_path, mode="r", encoding="utf-8") as file:
    reader = csv.reader(file)

    print("4-letter Cyrillic names:")
    for row in reader:
        if not row:  # skip empty rows
            continue
        cyrillic_name = row[0].strip()
        if len(cyrillic_name) == 4:
            print(f"{cyrillic_name} -> Mongolian: {row[1]}, Latin: {row[2]}")
```
